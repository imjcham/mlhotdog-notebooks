```python
import string
from bs4 import BeautifulSoup
import urllib2
import pandas as pd


#set variables
loop_lastname = list(string.ascii_uppercase)
#loop_lastname = ['A', 'B']
loop_year = map(str, list(range(2003,2018)))

site_url = "http://basketball.realgm.com/"
summary_url = "http://basketball.realgm.com/ncaa/players/"+ str(loop_year[0]) + "/" + loop_lastname[0]
player_url = ""

```

```python
#column headers
headers = []
url = "http://basketball.realgm.com/ncaa/players/2004/A"
content = urllib2.urlopen(url).read()
soup = BeautifulSoup(content)

#start scraping
#find first table tag
table = soup.find('table')

#find header tags to create columns and store in headers[]
summary_columns = table.findAll('th')
for x in summary_columns:
    headers.append(x.getText())
headers.append('Player URL')
```

```python
print headers
```

```python
#stats column data
summary_data = []
player_data = []
for year in loop_year:
  for name in loop_lastname:
    url = "http://basketball.realgm.com/ncaa/players/" + year + "/" + name
    #url = "http://basketball.realgm.com/ncaa/players/2007/X"
    print url


    #grab content
    content = urllib2.urlopen(url).read()
    soup = BeautifulSoup(content)

    #start scraping
    #find first table tag
    table = soup.find('table')
    if table:
      #grab basic stats and put into summary_data[]
      for r in table.findAll('tr'):
        data =[]
        for c in r.findAll('td'):
          data.append(c.getText())
        if data:
          data.append(r.find('a', href=True)['href'])
          summary_data.append(data)

  #print summary_data
  #put in pandas dataframe
  #print(len(headers))
df = pd.DataFrame(summary_data, columns=headers)
```
